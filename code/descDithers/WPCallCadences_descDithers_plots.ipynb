This notebook displays diagnostic plots saved by `descDithers_wp_cadences.py` for all the cadences included in the Cadence White Paper call; the outputs are read from `/global/homes/a/awan/desc/wp_descDithers_csvs` at NERSC.

```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image, display
import os
```

```python
outDir = '/global/homes/a/awan/desc/wp_descDithers_csvs'
```

```python
for png in [f for f in os.listdir(outDir) if f.endswith('png')]:
    display(Image(filename='%s/%s'%(outDir, png)) )
```

We see that the translational dithers retain the overall characteristics of the pointing RA, Dec, while the rotational dithers change the distribution rather significantly (for most cadences) from strongly skewed to a comparatively more uniform one.
