```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image, display
import os
```

```python
outdir = '/global/homes/a/awan/LSST/lsstRepos/ObsStrat/postwp/results/fom_figs/'
```

```python
for year in ['Y1', 'Y3', 'Y6', 'Y10']:
    for png in [f for f in os.listdir(outdir) if f.endswith('png') and \
                f.__contains__('noprior') and f.__contains__('contour') and \
                f.__contains__('%s_' % year) and not f.__contains__('limited')]:
        print(png)
        display(Image(filename='%s/%s'%(outdir, png)) )
```
